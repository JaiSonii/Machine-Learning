# Dealing With Duplicate Values

```python
import pandas as pd
```

```python
cities = ['New York', 'Los Angeles', 'Chicago', 'Austin', 'Phoenix', 'Los Angeles', 'Chicago']
rankings = [1,2,3,4,5,6,7]
data = [9.8, 9.5, 9.2, 8.7, 8.5, 8.2, 8.1]

dict = {'City' : cities, 'Ranking' : rankings, 'Rating': data}

data_frame = pd.DataFrame(dict)
```

```python
data_frame
```

```python
data_frame.duplicated()
```

```python
data_frame.duplicated().sum()
```

```python
data_frame.duplicated('City').sum()
```

```python
data_frame.drop_duplicates('City', keep = 'first')
```

- keep = last

```python
data_frame.drop_duplicates('City', keep = 'last')
```

- keep = False

```python
data_frame.drop_duplicates('City', keep = False)
```
