# Handling Null Values

```python
import pandas as pd
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

- ### Deleting the rows with null values

```python
df.shape
```

```python
df1 = df.dropna()
```

```python
df1.shape
```

```python
df1.isna().sum()
```

- ### Removing the Columns having null values

```python
df2 = df.copy()
```

```python
df2.tail()
```

```python
df2.isna().sum()
```

```python
df2.drop('last_review', axis = 1, inplace = True)
```

```python
df2.isna().sum()
```

- ### Filling the null values

```python
df3 = df.copy()
```

```python
df3.sample(6)
```

- Calculating the mean for reviews_per_month

```python
mean_value = round(df3['reviews_per_month'].mean(), 2)
mean_value
```

```python
df3['reviews_per_month'].fillna(mean_value, inplace = True)
```

```python
df3.sample(6)
```

- Calculating the mode for last_review

```python
mode_value = df3['last_review'].mode()[0]
mode_value
```

```python
df3['last_review'].fillna(mode_value, inplace = True)
```

```python
df3.sample(6)
```

- ### Categorical Data

```python
df4 = df.copy()
```

```python
df4.sample(5)
```

```python
df4.nunique()
```

```python
most_frequent = df4['last_review'].value_counts().index[0]
most_frequent
```

```python
df4['last_review'].fillna(most_frequent, inplace = True)
```

```python
df4.sample(10)
```
