- ## Data Analysis for Categorical Data

```python
df.info()
```

```python
df['id'] = df['id'].astype(str)
df['host_id'] = df['host_id'].astype(str)
```

```python
df['neighbourhood_group'].value_counts().reset_index().rename(columns = {'neighbourhood' : 'area', 'count' : 'number_of_hotels'})
```

`showing percentage`

```python
df['neighbourhood_group'].value_counts(normalize = True).reset_index().rename(columns = {'neighbourhood' : 'area'})
```

```python
df['neighbourhood'].value_counts()
```

```python
df['neighbourhood'].value_counts().reset_index().rename(columns = {'count' : 'number_of_hotels'})
```

```python
df_2 = df['neighbourhood'].value_counts().reset_index().rename(columns = {'count' : 'number_of_hotels'})
```

```python
df_2[df_2['number_of_hotels'] > 1000]
```

```python
df['room_type'].value_counts()
```
