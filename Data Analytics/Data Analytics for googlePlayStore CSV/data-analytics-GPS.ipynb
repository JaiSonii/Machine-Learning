- ### Removing the rows with null values

```python
df = data.dropna()
```

## Q1. Calculating the average values 

```python
df['Rating']
```

```python
rating_avg_value = df['Rating'].sum()/len(df['Rating'])
print('The average rating is', round(rating_avg_value,2))
```

## Q2. How many apps with rating 5.0

```python
## Approach 1
print(len(data[data['Rating'] == 5.0]))
```

```python
# Approach 2
count = 0
for i in data['Rating']:
    if i == 5.0:
        count += 1
print('There are', count, 'app with rating 5.0')
```

# Categorical Data Analysis

```python
data.head()
```

```python
data['Category'].unique()
```

## Q3. Number of Applications in each unique category

```python
category_dict = {}
for category in data['Category'].unique():
    count = 0
    for name in data['Category']:
        if name == category:
            count += 1
    category_dict[category] = count
    print(category, ': ', count)
```

```python
print(category_dict)
```

## Q4. Number of Free and Paid Applications

```python
data['Type'].unique()
```

```python
df['Type'].value_counts()
```

```python
# Another Manual Approach
free_count = 0
for i in df['Type']:
    if i == 'Free':
        free_count += 1
print('There are', free_count, 'free Applications')
```

```python
print('There are', len(df['Type']) - free_count, 'paid Applications')
```

- ### Percentage of free Apps


```python
prcnt_free = round(free_count/len(df['Type']) * 100, 2)
print('Percentage of free Apps: ', prcnt_free)
```

- ### Dict for Content Rating

```python
content_rating_dict = {}
for name in data['Content Rating'].unique():
    count = 0
    for rating in data['Content Rating']:
        if rating == name:
            count += 1
    content_rating_dict[name] = count
print(content_rating_dict)
```

## Null Values Handling using Imputer

```python
import numpy as np
from sklearn.impute import SimpleImputer
import pandas as pd
```

```python
data = pd.read_csv('googleplaystore.csv')
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
impute.fit(data.iloc[:, 2:3].values)
```

```python
data.iloc[:, 2:3] = impute.transform(data.iloc[:, 2:3].values).round(1)
```

```python
data.isnull().sum()
```
