# in hierarchy

```python
os.getcwd()
```

```python
os.chdir("../")
```

```python
os.getcwd()
```

```python
master="master"
os.mkdir(master)
m=3
n=2
for i in range(m):
    path=master + "/" + "child" + str(i+1)
    os.mkdir(path)
    for j in range(n):
        path=master + "/" + "child" + str(i+1) + "/" + "grand_child" + str(j+1)
        os.mkdir(path)
print(m + (m*n) + 1)        
```

# reading bulk files

```python
os.mkdir("files")
```

```python
files = os.listdir("files")
for i in files:
    if i!='.ipynb_checkpoints':
        op = open("files/" + i , "r" )
        print(f"Content in files/{i}      :   {op.read()} " )
```

# writing bulk file text in one file

```python
os.getcwd()
```

```python
files = os.listdir("files")
fd_write=open("data.txt" , "w")
for i in files:
    if i!='.ipynb_checkpoints':
        fd_read = open("files/" + i , "r" )
        fd_write.write(fd_read.read() + "\n")
fd_write.close()        
```

```python
fd=open("data.txt","r")
print(fd.read())
fd.close()
```
