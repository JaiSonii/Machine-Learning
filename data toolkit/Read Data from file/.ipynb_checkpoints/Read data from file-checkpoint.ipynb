# Text PreProcessing

```python
rd = open('text.txt', 'r', encoding= 'utf-8')
data = rd.read()
print(data)
rd.close()
```

```python
paragraphs = data.split('\n\n')
print('No of paragraphs in the file is : ', len(paragraphs))
```

```python
words = data.split()
print('Total number of words in the file are: ',len(words))
```

```python
lines = data.split('.')
print('The total number of lines in the file is: ', len(lines[:-1]))
```

```python
for i in range(7,32):
    pattern = '[' + str(i) +']'
    data = data.replace(pattern,'')
```

```python
print(data)
```

# Creating and Appending

## create

```python
fd = open('new.txt', 'w')
fd.write('This is a new file and some content is written into it.')
fd.close()
```

## append

```python
fd = open('new.txt', 'a')
fd.write('This is the appended text in the new text file')
fd.close()
```

## append in new line

```python
fd = open('new.txt', 'a')
txt = input("Enter the text")
txt = txt + '.\n'
fd.write(txt)
fd.close()
```

```python
fd = open('new.txt' , 'r')
print(fd.read())
```
