```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

# sin and cos graphs

```python
x = np.linspace(0, 2*np.pi, 50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label = 'sine')
plt.plot(x, cos_y, label = 'cosine')

plt.legend(loc = 'upper right')
plt.title('Sine and Cosine')

plt.show()
```

## Dashed Format

```python
plt.plot(x, sin_y, '*')
plt.plot(x, cos_y, '--')

plt.title('Sine and Cosine')
plt.show()
```
