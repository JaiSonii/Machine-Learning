```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

# sin and cos graphs

```python
x = np.linspace(0, 2*np.pi, 50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label = 'sine')
plt.plot(x, cos_y, label = 'cosine')

plt.legend(loc = 'upper right')
plt.title('Sine and Cosine')

plt.show()
```

## Dashed Format

```python
plt.plot(x, sin_y, '*')
plt.plot(x, cos_y, '--')

plt.title('Sine and Cosine')
plt.show()
```

## Applying Scatter Plot on DataSet

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = iris['data']
x = data[:,:1]
y = data[:,1:2]

plt.scatter(x,y)
plt.title('Scatter Plot of Sepal Length and Sepal Width')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])

plt.show()
```

# Adding Markers

```python
plt.scatter(x,y, marker = 'x')
plt.title('Scatter Plot of Sepal Length and Sepal Width')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.show()
```

## cmap and edge colors

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = rng.rand(100) * 1000

plt.scatter(x , y, c = colors, s = sizes, cmap = 'viridis', edgecolors= 'black', alpha = 0.7)
plt.title('Using cmap and edgecolors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.colorbar()
plt.show()
```

# using cmap and others on iris dataset

```python
iris = load_iris()
data = iris['data']
x = data[:,:1]
y = data[:,1:2]

plt.scatter(x,y, alpha = 0.7, s = data[:,3] * 100, c = iris.target, cmap = 'viridis', edgecolor = 'black')
plt.colorbar()

plt.title('Scatter Plot of Sepal Length and Sepal Width')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])

plt.show()
```

```python
# histogram types

x = np.random.randn(10000)

plt.hist(x, edgecolor = 'black', bins = 100, color = 'red', alpha = 0.8, histtype = 'stepfilled')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

## Multiple Histograms on same plot

```python
x1 = np.random.randn(100)
x2 = np.random.randn(100) + 3
x3 = np.random.randn(100) + 6

plt.hist(x1, bins = 10, edgecolor = 'black', alpha = 0.8, histtype = 'barstacked')
plt.hist(x2, bins = 10, edgecolor = 'lightblue', alpha = 0.8, histtype = 'barstacked')
plt.hist(x3, bins = 10, edgecolor = 'white', alpha = 0.8, histtype = 'barstacked')

plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Multiple Histograms')

plt.show()
```

## 2D Histograms

# pie charts

```python
cars = ['AUDI','BMW','MERCEDES','VOLKSWAGEN','TESLA','TOYOTA']
data = [38,50,15,20,1,60]

plt.pie(data, labels = cars )
plt.show()
```

- ### Basic operations

```python
explode = [0,0,0,0,0.1,0]
plt.pie(data, labels = cars, explode= explode )
plt.show()
```

```python
explode = [0,0,0,0,0.1,0]
plt.pie(data, labels = cars, explode= explode, autopct = '%2.1f%%' )
plt.show()
```

```python
explode = [0.1,0.1,0.1,0.1,0.2,0.1]
plt.pie(data, labels = cars, explode= explode, autopct = '%2.1f%%', shadow = True )
plt.show()
```

```python
plt.style.use('tableau-colorblind10')
explode = [0.1,0.1,0.1,0.1,0.2,0.1]
plt.pie(data, labels = cars, explode= explode, autopct = '%2.1f%%', shadow = True, wedgeprops = {'edgecolor' : 'white', 'linewidth' : 0.5} )
plt.show()
```

```python
plt.style.use('tableau-colorblind10')
explode = [0.1,0.1,0.1,0.1,0.2,0.1]
plt.pie(data, labels = cars, explode= explode, autopct = '%2.1f%%', shadow = True, wedgeprops = {'edgecolor' : 'white', 'linewidth' : 0.5, 'antialiased' : True })
plt.show()
```

# Multiple Subplots

```python
plt.style.use('tableau-colorblind10')
x = np.linspace(0, 4*np.pi, 50)

fig, ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sine')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cosine')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan')


y = np.random.randn(100)
ax[1,1].hist(y)
ax[1,1].set_title('Random Distribution')

plt.tight_layout()
plt.show()
```

# `Example 3`

```python
x = np.linspace(0, 4*np.pi, 50)
y = np.sin(x)

fig, (ax1,ax2, ax3) = plt.subplots(1,3, figsize = (15,5))

ax1.plot(x,y)
ax2.plot(x, y**2)
ax3.plot(x, y**3)

ax1.set_title('sin(x)')
ax2.set_title('sq(sin(x))')
ax3.set_title('cube(sin(x))')
```
