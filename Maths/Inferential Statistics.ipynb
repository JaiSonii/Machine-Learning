# Hypothesis Testing

- ## Ztest

```python
import numpy as np
import scipy.stats as st
```

```python
alpha = 0.05
n = 50
sample_mean = 110
pop_mean = 100
pop_sd = 15
```

```python
def ztest(n, sample_mean, pop_mean, pop_sd, alternate):
    z = (sample_mean - pop_mean)/(pop_sd/n**0.5)
    p_value = st.norm.sf(z)
    print(f'z-value : {z} \np-value : {p_value}')

    if(alternate == 'one-sided'):
        if(p_value < alpha):
            print('Reject the Null Hypothesis')
        else:
            print('Failed to reject the Null Hypothesis')
    elif(alternate == 'two-sided'):
        if z < alpha/2 :
            print('Reject the Null Hypothesis')
        else:
            print('Failed to reject the Null Hypothesis')
```

```python
ztest(n, sample_mean, pop_mean, pop_sd, 'one-sided')
```

```python
# example 2 for ztest
alpha = 0.05
n = 81
sample_mean = 90
pop_mean = 82
pop_sd = 20
```

```python
ztest(n, sample_mean, pop_mean, pop_sd, 'one-sided')
```

- ## Ttest

```python
# data 
sample_mean = 140
pop_sd = 20
alpha = 0.05
data = pop_sd* np.random.randn(30) + sample_mean
print(np.mean(data), np.std(data))
```

```python
t_statistic, p_value = st.ttest_1samp(a = data, popmean= 100)
```

```python
print(f't_statistic : {t_statistic}\np_value : {p_value}')
```

```python
if(p_value < alpha):
    print('Reject the Null Hypothesis')
else: 
    print('Failed to reject the Null Hypothesis')
```

- ## Chi-Squared Test

```python
from scipy.stats import chisquare
```

`The chisquare() function from the scipy.stats library helps to calculate the p_value and chi-squared value for the catogorical data and compare the expected and observed results`

The example used here has values 48,24,24 are the observed values of the subjects being favorable independent or unbiased, and the expected values are taken to be 32,32,32, Now on applying the chisquare function 

- - ### Example 1

```python
chi_value, p_value = chisquare([48,24,24])
```

```python
print(f'chi-value : {chi_value}\np-value : {p_value}')
```

`Now this value can be compared with the alpha which is given to be equals to 0.05`

```python
if p_value < alpha :
    print('Dependent (Reject H0)')
else: 
    print('Independent and (H0 holds true)')
```

`This implies that Reject the Null Hypothesis`

 - - ### Example 2

`In this example we test the bolts manufactuered from 5 machines and the results of defective bolts are 5,9,13,7,6 and expected are 8,8,8,8,8`

```python
chi_value, p_value = chisquare([5,9,13,7,6])
```

```python
print(f'chi-value : {chi_value}\np-value : {p_value}')
if p_value < alpha :
    print('Dependent (Reject H0)')
else: 
    print('Independent and (H0 holds true)')
```

- ## Anova

- - ### 1 Way Anova

- - - #### Example 1

```python
from scipy.stats import f_oneway
```

```python
english = [2,4,2]
maths = [2,3,4]
science = [1,2,5]
```

```python
alpha = 0.05
```

```python
stat, p_value = f_oneway(english, maths , science)
```

```python
print(f'Stat : {stat}\np-value : {p_value}')
```

```python
if p_value < alpha :
    print('Reject the Null Hypothesis: At least one of them is different')
else:
   print('Failed to Reject Null Hypothesis: There is no difference') 
```

- - - #### Example 2

- - ### 2 Way Anova

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
data_frame = pd.DataFrame({
    'Fertilizer': np.repeat(['daily', 'weekly'], 15),
    'Watering' : np.repeat(['daily', 'weekly'], 15),
    'Height' : [14,16,15,15,16,13,12,11,14,15,16,16,17,18,14,13,14,14,14,15,16,16,17,18,14,13,14,14,14,15]
})
```

```python
data_frame
```

```python
model = ols("Height~ Fertilizer + Watering + Fertilizer:Watering", data_frame).fit()
```

```python
result = sm.stats.anova_lm(model, type = 2)
```
