```python
import pandas as pd
from scipy.stats import binom 
```

`Calculating the probability of number of questions correct from 20 questions using probability mass function`

```python
num_of_questions = 20
prob_correct =0.25
```

```python
questions = [i for i in range(1,num_of_questions+1)]
```

```python
df = pd.DataFrame(questions)
```

```python
df = df.rename(columns = {0 : 'num_of_correct_ques'})
```

`Defining the lambda function for probability calculation`

```python
probability = lambda x : binom.pmf(x, num_of_questions, prob_correct)
```

```python
df
```

```python
df['probability'] = df['num_of_correct_ques'].apply(probability)
```
