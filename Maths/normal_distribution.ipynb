# Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
```

```python
x = np.arange(-5,5, 0.00001)
```

```python
x
```

```python
y_data = stats.norm.pdf(x, 0, 1)
```

```python
y_data
```

```python
plt.plot(x, y_data)
```

`Increasing the variation and mean`

```python
y_data = stats.norm.pdf(x, 2, 2)
fig = plt.figure(figsize = (15,8))
plt.plot(x, y_data)
```
