```python
class graph:
    def __init__(self):
        self.adj = {}

    def addEdge(self, vertex, edge, direction):
        if vertex not in self.adj:
            self.adj[vertex] = []
        self.adj[vertex].append(edge)
        if direction == 0:
            if edge not in self.adj:
                self.adj[edge] = []
            self.adj[edge].append(vertex)

    def printGraph(self):
        for key in self.adj.keys():
            print(f'{key} -> {self.adj[key]}')

g = graph()
g.addEdge(1,3,0)
g.addEdge(1,2,0)
```

```python
g.printGraph()
```
