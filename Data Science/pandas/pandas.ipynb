# Append

```python
ser1 = pd.Series([1,2,5])
ser2 = pd.Series([7,9,8])
```

# All Strings

```python
ser = pd.Series(['  indore   ',' bhopal', ' chhindwara   ', '   seoni', ' ujjain'])
for i in ser:
    print(f"{i} -- {len(i)}")
```

```python
ser.str.upper()
```

```python
# Removing Spaces

ser = ser.str.strip()

for i in ser:
    print(f'{i} -- {len(i)}')
```

```python
ser = ser.str.upper()
print(ser)
```

```python
ser.str.lower()
```
