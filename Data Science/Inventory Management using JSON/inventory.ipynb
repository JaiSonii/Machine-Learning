```python
import json
import time

sale = ''

fd = open('Inventory.json', 'r')
js = fd.read()
fd.close()

products = json.loads(js)

print('-'*13 + 'MENU' + '-'*13)
for key in products.keys():
    print(key, '  |  ', products[key]['Name'], '|', products[key]['Price'], '|', products[key]['Quantity'])

ui_name = input('Enter your name: ')
ui_phone = input('Enter yout phone: ')
ui_product_id = str(input('Enter the Product ID: '))
ui_product_qn = int(input('Enter the quantity: '))

final_qn = ui_product_qn

if products[ui_product_id]['Quantity'] == 0:
    print('-'*40)
    print('Sorry for the inconvinience, We are out of stock')

elif ui_product_qn > products[ui_product_id]['Quantity']:
    print('-'*40)
    print(f"Sorry for the inconvenience, \nWe are only having {products[ui_product_id]['Quantity']} {products[ui_product_id]['Name']}")
    ch = input('Press Y/y to purchase: ')
    if ch == 'Y' or ch == 'y':
        final_qn = products[ui_product_id]['Quantity']
        total_cost = products[ui_product_id]['Price'] * final_qn
        print('-'*40)
        print(' '*18 + 'Bill' + ' '*18)
        print('Name               :  ', products[ui_product_id]['Name'])
        print('Price              :  ', products[ui_product_id]['Price'], 'Rs')
        print('Quantity           :  ', final_qn)
        print('-' * 40)
        print('Total              :  ', total_cost, 'Rs')
        print('-' * 40)
        products[ui_product_id]['Quantity'] -= final_qn

        sale = ui_name + ',' + ui_phone + ',' + ui_product_id + ',' + str(final_qn) + ',' + str(products[ui_product_id]['Price']) + ',' + str(total_cost) + ',' + time.ctime() + '\n'
        
    
    else:
        print('-'*40)
        print('Thanks')
else:
    total_cost = products[ui_product_id]['Price'] * final_qn
    print('-'*40)
    print(' '*18 + 'Bill' + ' '*18)
    print('Name               :  ', products[ui_product_id]['Name'])
    print('Price              :  ', products[ui_product_id]['Price'] , 'Rs')
    print('Quantity           :  ', final_qn)
    print('-' * 40)
    print('Total              :  ', total_cost, 'Rs')
    print('-' * 40)
    products[ui_product_id]['Quantity'] = products[ui_product_id]['Quantity'] - ui_product_qn
    products[ui_product_id]['Quantity'] -= final_qn

    sale = ui_name + ',' + ui_phone + ',' + ui_product_id + ',' + str(final_qn) + ',' + str(products[ui_product_id]['Price']) + ',' + str(total_cost) + ',' + time.ctime() + '\n'

fd = open('sales.txt', 'a')
fd.write(sale)
fd.close()

fd = open('Inventory.json', 'w')
fd.write(json.dumps(products))
fd.close()

```
